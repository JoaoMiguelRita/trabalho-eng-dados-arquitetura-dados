```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = (
    SparkSession.builder
    .appName("DeltaExample")
    .master("local[*]")
)

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
from pyspark.sql.types import *
from decimal import Decimal

data = [
    (1, "KOHI 09QC 1HV", "KOMECO", Decimal("1798.90"), 2023, "9000 BTU/h"),
    (2, "B0DSLQP69S", "SAMSUNG", Decimal("2089.10"), 2024, "12000 BTUs"),
    (3, "42AGVCC30M5", "SAMSUNG", Decimal("6590.00"), 2025, "30000 BTUs")
]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("nm_modelo", StringType(), True),
    StructField("nm_marca", StringType(), True),
    StructField("valor", DecimalType(10,2), True),
    StructField("ano", IntegerType(), True),
    StructField("potencia", StringType(), True)
])

df = spark.createDataFrame(data, schema)
df.show()
```

```python
( df.write
  .format("delta")
  .mode("overwrite")
  .save("./RAW/ar_condicionado_delta")
)
```

```python
df_delta = spark.read.format("delta").load("./RAW/ar_condicionado_delta")
df_delta.show()
```

```python
spark.sql("SELECT * FROM ar_condicionado_delta").show()
```

```python
from delta.tables import DeltaTable

arCondicionado = DeltaTable.forPath(spark, "./RAW/ar_condicionado_delta")
```

```python
arCondicionado.history().show()
```

```python
spark.sql(
    """
    INSERT INTO ar_condicionado_delta VALUES 
    (1, 'KOHI 09QC 1HV', 'KOMECO', 1798.90, 2023, '9000 BTU/h'),
    (2, 'B0DSLQP69S', 'SAMSUNG', 2089.10, 2024, '12000 BTUs'),
    (3, '42AGVCC30M5', 'SAMSUNG', 6590.00, 2025, '30000 BTUs')
    """
)
```

```python
spark.sql("select * from ar_condicionado_delta").show()
```

```python
arCondicionado.history().show(truncate=False)
```

```python
spark.sql(
    """
    ALTER TABLE ar_condicionado_delta 
      ADD COLUMNS (fl_inverter BOOLEAN, tensao STRING)
    """
)
```

```python
spark.sql(
    """
    select * from ar_condicionado_delta
    """
).show()
```

```python
arCondicionado.update(
    condition="id IN (2, 3)",
    set={
        "fl_inverter": "true",
        "tensao": "'220v'"
    }
)
```

```python
spark.sql(
    """
    select * from ar_condicionado_delta
    """
).show()
```

```python
DeltaTable.isDeltaTable(spark, "spark-warehouse/ar_condicionado_delta")
```

```python
spark.sql('select * from ar_condicionado_delta').show()
```

```python
arCondicionado.update(
    condition="id = 1",
    set={
        "fl_inverter": "false",
        "tensao": "'170v'"
    }
)
```

```python
spark.sql('select * from ar_condicionado_delta').show()
```
